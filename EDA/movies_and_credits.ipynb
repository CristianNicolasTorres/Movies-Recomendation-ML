```python
import pandas as pd
```

```python
df_movies=pd.read_csv('D:\SoyHenry\Proyecto Integrador 1\Movies-Recomendation-ML\Data\movies_dataset_ETLready.csv')
df_credits=pd.read_csv('Data\credits_ETLready.csv')
```

```python
merged_df = pd.merge(df_movies, df_credits, on="id")
merged_df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select the columns of interest
columns = ['budget', 'revenue', 'release_year', 'popularity', 'vote_count', 'vote_average']

# Create a correlation matrix
correlation_matrix = merged_df[columns].corr()

# Crear a heat map
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Mapa de Calor - Relación entre Variables')
plt.show()
```

```python
columns_to_drop = ['overview', 'runtime', 'status', 'tagline', 'collection_id',
                   'collection_name', 'production_companies_extracted', 'production_countries_extracted']

merged_df.drop(columns=columns_to_drop, axis=1, inplace=True)
```
