```python
import pandas as pd
import ast
url='https://drive.google.com/file/d/1lMGJUWVVVRPO00ZWqzEJZIxFhSqtfmAB/view?usp=drive_link'
url = 'https://drive.google.com/uc?id=' + url.split('/')[-2] 
df_credits=pd.read_csv(url)
```

```python
df_credits['crew']
```

```python
def extract_cast(row):
    try:
        casts_list = ast.literal_eval(row)
        casts = [cast['name'] for cast in casts_list]
        return casts
    except (SyntaxError, ValueError):
        return []

df_credits['casts_extracted'] = df_credits['cast'].apply(extract_cast)
df_credits
```

```python
def extract(row):
    try:
        liste = ast.literal_eval(row)
        director_names = [new_row['name'] for new_row in liste if new_row['job'] == 'Director']
        return director_names
    except (SyntaxError, ValueError):
        return []


df_credits['directors_extracted']=df_credits['crew'].apply(extract)
```

```python
df_credits
```

```python
df_removed_columns_credits = pd.DataFrame()
list_to_drop=['crew','cast']
df_removed_columns_credits[list_to_drop] = df_credits[list_to_drop]
df_credits.drop(list_to_drop, axis=1, inplace=True)
```

```python
df_credits.drop_duplicates(subset="id", inplace=True)
```
